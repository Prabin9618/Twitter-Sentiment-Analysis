```python
text = 'cricket'
```

```python
tweets_list = []
```

```python
for status in tweepy.Cursor(api.search_tweets,q=text,lang='en',result_type='recent').items(500):
        tweets_list.append(status.text)
```

```python
tweets_list
```

```python
final_list = []
for i in tweets_list:
  tweet = i.lower()
  tweet = re.sub(r'\W',' ',tweet)
  tweet = re.sub(r'\s+',' ',tweet)
  tweet = re.sub('[^a-zA-Z]',' ',tweet)
  final_list.append(tweet)
```

```python
final_list
```

```python
with open('model.pickle','rb') as file:
    model = pickle.load(file)

with open('analyzer.pickle','rb') as file:
    analyzer = pickle.load(file)
```

```python
positive = 0
negative = 0
neutral = 0

for tweet in final_list:
  sentiment = model.predict(analyzer.transform([tweet]).toarray())
  if sentiment[0] == 1:
    positive += 1
  elif sentiment[0] == -1:
    negative += 1
  else:
    neutral += 1
```
